# Gradient-boost

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame()
```

```python
df['X'] = X.reshape(100)
df['y'] = y 
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.title('X vs Y')
```

```python
df['pred1'] = df['y'].mean()
```

```python
df
```

```python
df['res1'] = df['y'] - df['pred1']
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.plot(df['X'],df['pred1'],color='red')
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree1 = DecisionTreeRegressor(max_leaf_nodes=8)
# 8 to 32 it is a good result 
```

```python
tree1.fit(df['X'].values.reshape(100,1),df['res1'].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1)
plt.show()
```

```python
# generating X_test
X_test = np.linspace(-0.5, 0.5, 500)
```

```python
y_pred = 0.265458 + tree1.predict(X_test.reshape(500, 1))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2, color='red')
plt.scatter(df['X'],df['y'])
```

```python
df['pred2'] = 0.265458 + tree1.predict(df['X'].values.reshape(100,1))
```

```python
df
```

```python
df['res2'] = df['y'] - df['pred2']
```

```python
df
```

```python
tree2 = DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree2.fit(df['X'].values.reshape(100,1),df['res2'].values)
```

```python
y_pred = 0.265458 + sum(regressor.predict(X_test.reshape(-1, 1)) for regressor in [tree1, tree2])
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
plt.title('X vs y')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

def gradient_boost(X, y, n_estimators=5, lr=0.1, regs=None, y_pred=None, plot=True):
    # initialize model storage and predictions
    if regs is None:
        regs = []
        y_pred = np.zeros_like(y)
    
    # base case
    if len(regs) >= n_estimators:
        return regs
    
    # compute residuals (errors)
    residuals = y - y_pred
    
    # fit weak learner
    tree = DecisionTreeRegressor(max_depth=3, random_state=42)
    tree.fit(X, residuals)
    
    # update ensemble prediction
    y_pred += lr * tree.predict(X)
    regs.append(tree)
    
    # --- plotting section ---
    if plot:
        x1 = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)
        y_boost = np.zeros_like(x1.flatten())
        for reg in regs:
            y_boost += lr * reg.predict(x1)
        
        plt.figure(figsize=(6, 4))
        plt.scatter(X, y, color='r', label='True')
        plt.plot(x1, y_boost, 'b-', linewidth=2, label=f'Step {len(regs)}')
        plt.title(f'Gradient Boosting - Iteration {len(regs)}')
        plt.legend()
        plt.show()
    
    # recursive call
    return gradient_boost(X, y, n_estimators, lr, regs, y_pred, plot)

# --- Example usage ---
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)

gradient_boost(X, y, n_estimators=5, lr=0.5)
```
